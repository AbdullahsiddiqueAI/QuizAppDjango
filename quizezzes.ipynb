```python
import random
quiz_bank = [
    {
        'question': 'What is the capital of Japan?',
        'options': ['Beijing', 'Tokyo', 'Seoul', 'Bangkok'],
        'answer': 'Tokyo'
    },
    {
        'question': 'Which element has the chemical symbol "O"?',
        'options': ['Oxygen', 'Gold', 'Silver', 'Iron'],
        'answer': 'Oxygen'
    },
    {
        'question': 'What is the largest ocean on Earth?',
        'options': ['Atlantic Ocean', 'Indian Ocean', 'Southern Ocean', 'Pacific Ocean'],
        'answer': 'Pacific Ocean'
    },
    {
        'question': 'Who wrote "Romeo and Juliet"?',
        'options': ['Charles Dickens', 'William Shakespeare', 'Jane Austen', 'Mark Twain'],
        'answer': 'William Shakespeare'
    },
    {
        'question': 'What is the capital of Brazil?',
        'options': ['Brasília', 'Rio de Janeiro', 'São Paulo', 'Buenos Aires'],
        'answer': 'Brasília'
    },
    {
        'question': 'In which year did the World War II end?',
        'options': ['1942', '1945', '1950', '1939'],
        'answer': '1945'
    },
    {
        'question': 'Which planet is known as the "Red Planet"?',
        'options': ['Earth', 'Mars', 'Venus', 'Jupiter'],
        'answer': 'Mars'
    },
    {
        'question': 'What is the currency of Germany?',
        'options': ['Dollar', 'Euro', 'Pound', 'Yen'],
        'answer': 'Euro'
    },
    {
        'question': 'Who painted the Mona Lisa?',
        'options': ['Vincent van Gogh', 'Leonardo da Vinci', 'Pablo Picasso', 'Michelangelo'],
        'answer': 'Leonardo da Vinci'
    },
    {
        'question': 'What is the capital of Australia?',
        'options': ['Sydney', 'Melbourne', 'Canberra', 'Brisbane'],
        'answer': 'Canberra'
    },
    {
        'question': 'Which is the largest mammal?',
        'options': ['Elephant', 'Blue Whale', 'Giraffe', 'Hippopotamus'],
        'answer': 'Blue Whale'
    },
    {
        'question': 'Who wrote "To Kill a Mockingbird"?',
        'options': ['Harper Lee', 'J.K. Rowling', 'George Orwell', 'Ernest Hemingway'],
        'answer': 'Harper Lee'
    },
    {
        'question': 'What is the main ingredient in guacamole?',
        'options': ['Tomato', 'Avocado', 'Onion', 'Lime'],
        'answer': 'Avocado'
    },
    {
        'question': 'What is the speed of light?',
        'options': ['300,000 km/s', '150,000 km/s', '600,000 km/s', '900,000 km/s'],
        'answer': '300,000 km/s'
    },
    {
        'question': 'Which continent is known as the "Dark Continent"?',
        'options': ['Asia', 'Africa', 'South America', 'Antarctica'],
        'answer': 'Africa'
    },
    {
        'question': 'What is the capital of Canada?',
        'options': ['Toronto', 'Ottawa', 'Vancouver', 'Montreal'],
        'answer': 'Ottawa'
    },
    {
        'question': 'Who developed the theory of relativity?',
        'options': ['Isaac Newton', 'Galileo Galilei', 'Albert Einstein', 'Stephen Hawking'],
        'answer': 'Albert Einstein'
    },
    {
        'question': 'What is the national flower of Japan?',
        'options': ['Rose', 'Tulip', 'Cherry Blossom', 'Lily'],
        'answer': 'Cherry Blossom'
    },
    {
        'question': 'Which gas is most abundant in the Earth\'s atmosphere?',
        'options': ['Oxygen', 'Nitrogen', 'Carbon Dioxide', 'Argon'],
        'answer': 'Nitrogen'
    },
    {
        'question': 'Who wrote "The Great Gatsby"?',
        'options': ['F. Scott Fitzgerald', 'Jane Austen', 'Charles Dickens', 'Herman Melville'],
        'answer': 'F. Scott Fitzgerald'
    }
]

def generate_unique_numbers(n, start, end):
    unique_numbers = set()

    while len(unique_numbers) < n:
        unique_numbers.add(random.randint(start, end))

    return list(unique_numbers)


def conduct_quiz(quiz_bank):
    score = 0

    for i, question in enumerate(quiz_bank,1):
        print(f"\nQuestion {i}: {question['question']}")
        for j, option in enumerate(question['options'],1):
            print(f"{j}. {option}")

        user_answer = input("Your answer (enter the number): ").strip()

        try:
            user_answer = int(user_answer)
            if 1 <= user_answer <= len(question['options']):
                user_answer = question['options'][user_answer - 1]
            else:
                print("Invalid input. Skipping question.")
                continue
        except ValueError:
            print("Invalid input. Skipping question.")
            continue

        if user_answer == question['answer']:
            print("Correct!\n")
            score += 1
        else:
            print(f"Wrong! The correct answer is: {question['answer']}\n")

    return score
random_numbers = generate_unique_numbers(5, 0, 19)
print("Generated Unique Random Numbers:", random_numbers)
# filter_Quiz=[]
# for i in random_numbers:
#     filter_Quiz.append(quiz_bank[i])
# print("Welcome to the Quiz!\n")
# user_score = conduct_quiz(filter_Quiz)
# print(f"\nQuiz Completed! Your Score: {user_score}/{len(random_numbers)}")
```
